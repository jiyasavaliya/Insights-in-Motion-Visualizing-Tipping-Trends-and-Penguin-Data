DataFrame is a 2-dimensional labeled data structure from the pandas library that is used to hold data in tabular form. 
the head() method returns the first few rows of the DataFrame. By default, it shows the first 5 rows, but you can specify a different number if needed. 

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

x = [1,2,3,4,5,6,7]
y = [20,40,30,40,30,50,30]

df = pd.DataFrame({"Days":x, "No of people": y})
df.head(4)
```

seaborn itself doesn't have a show() function. You need to use matplotlib.pyplot.show() to display plots created with Seaborn. 

```python
sns.lineplot(x = "Days", y = "No of people", data = df)
plt.show()
```

```python
var1 = np.linspace(1,10,10).reshape(2,5)
var1
```

In Seaborn, annot is a parameter used in the heatmap function to add numerical annotations to each cell in the heatmap.

annot_kws is a parameter in Seaborn's heatmap function that allows you to customize the appearance of the annotations

cbar=True to display the color bar.

```python
y={"fontsize":20, "color":"r"}
v = sns.heatmap(var1, vmin=0, vmax=10, cmap='gist_heat', annot=True, annot_kws=y,
            linewidth=10, linecolor="#d62728", cbar=True, xticklabels=False, yticklabels=False)

v.set(xlabel="Seaborn", ylabel="JupyterNotebook")
sns.set(font_scale=0.4)
plt.show()
```

Using kind="reg" adds regression lines to the scatter plots, helping visualize the linear
relationship between variables. It makes trends or correlations easier to spot.

kind="kde" = it creates a smooth curve that shows how data points are distributed.

kind='hist' creates a histogram. 

kind='scatter' creates a scatter plot

Diagonal = Distribution of single variables (KDE/histograms).
Off-diagonal = Pairwise relationships (scatter plots) for each pair of variables.

```python
sns.pairplot(var2,hue="sex", hue_order=["Female", "Male"], kind='scatter', ) # 'scatter', 'kde', 'hist', 'reg'
plt.show()
```

```python
sns.pairplot(var2,hue="sex", hue_order=["Female", "Male"], kind='kde', diag_kind="hist") # 'scatter', 'kde', 'hist', 'reg'
plt.show()
```

```python
sns.pairplot(var2,hue="sex", hue_order=["Female", "Male"], kind='hist') # 'scatter', 'kde', 'hist', 'reg'
plt.show()
```

```python
sns.pairplot(var2,hue="sex", hue_order=["Female", "Male"], kind='reg') # 'scatter', 'kde', 'hist', 'reg'
plt.show()
```

styling plot


-figure styles
-removing axes spines

style = notebook, paper, poster

```python
sns.set_style("darkgrid")
sns.set_context("poster", font_scale=0.9)
sns.barplot(x="day", y="total_bill", data=var2, palette="cool")
plt.show()
```

```python
sns.set_style("white")
sns.barplot(x="day", y="total_bill", data=var2)
sns.despine()
plt.show()
```

```python
sns.set_style("white")
plt.figure(figsize=(12,5))
sns.barplot(x="day", y="total_bill", data=var2)
sns.despine()
plt.show()
```
